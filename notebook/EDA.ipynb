```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer

df = load_breast_cancer()
df
```

```python
d = pd.DataFrame(df.data, columns = df.feature_names)
d['target'] = df.target
d.head()
```

```python
d['target'].value_counts()
```

- Somewhat balanced dataset so no upsampling or downsampling needed

```python
d.describe().T
```

```python
d.isna().sum()
```

- No missing datapoint found

```python
import seaborn as sns

for i in list(d.columns):
    sns.boxplot(d[i])
    plt.show()
```

- Lots of unreplaceable outliers.

```python
# splitting dataset into independent and dependent features.
X = d.drop('target',axis=1)
y = d['target']
```

```python
# Split dataset into train and test dataset
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, random_state = 69)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score
```

```python
def model(X_train,X_test,y_train,y_test, CLASSIFIER):

    classifier = CLASSIFIER
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    return accuracy_score(y_test, y_pred) * 100
```

```python
classifiers = {
    'LogisticRegression' : model(X_train,X_test,y_train,y_test, LogisticRegression()),
    'DecisionTreeClassifier' : model(X_train,X_test,y_train,y_test, DecisionTreeClassifier()),
    'RandomForestClassifier' : model(X_train,X_test,y_train,y_test, RandomForestClassifier()),
    'SVC' : model(X_train,X_test,y_train,y_test, SVC()),
    'XGBClassifier' : model(X_train,X_test,y_train,y_test, XGBClassifier()),
    'GaussianNB' : model(X_train,X_test,y_train,y_test, GaussianNB())
}

for i,j in classifiers.items():
    print(f"{i} : {j}")
```

- So Random Forest Gave us the best model so we are going to pickle random forest model

```python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred) * 100
```

```python
## Hyperparameter Tuning

from sklearn.model_selection import GridSearchCV

p = {
    'max_depth':[3,5,10,15],
    'n_estimators':[50,100,150,200,300],
    'criterion':['gini','entropy']
}

cv = GridSearchCV(classifier, param_grid=p, scoring='accuracy', cv=5, verbose=3)
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
classifier = RandomForestClassifier(criterion='gini', max_depth=10, n_estimators=50)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
accuracy_score(y_test, y_pred) * 100
```

```python
import pickle

# Save the model to a file using pickle
with open('classifier.pkl', 'wb') as f:
    pickle.dump(classifier, f)
```
